```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0,2*np.pi,100)
print(x[-1],np.pi*2)

y = np.sin(x)
z = np.cos(x)
w = np.sin(4*x)
v = np.cos(4*x)

fig = plt.figure(figsize=(6,6))

plt.plot(x,y,label=r'$\sin(x)$')
plt.plot(x,z,label=r'$\cos(x)$')
plt.plot(x,w,label=r'$\sin(4x)$')
plt.plot(x,y,label=r'$\cos(4x)$')

plt.xlabel(r'$x$')
plt.ylabel(r'$y(x)$')
plt.xlim([0,np.pi*2])   
plt.ylim([-1.2,1.2])
plt.legend(loc=1,framealpha=0.95)
           
plt.gca().set_aspect(np.pi/1.2)
```
